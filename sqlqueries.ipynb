```python
import pandas as pd
df = pd.read_csv("cleaned_student_dataset.csv")
```

```python
import sqlite3
conn=sqlite3.connect("students.db")
df.to_sql("students",conn,if_exists="replace",index=False)
```

```python
query = """
SELECT COUNT(*) AS total_students
FROM students;
"""

pd.read_sql(query, conn)
```

```python
query="""
SELECT Gender, COUNT(*) AS total
FROM students
GROUP BY Gender;
"""
pd.read_sql(query,conn)
```

```python
query="""
SELECT Course, COUNT(*) AS total_students
FROM students
GROUP BY Course
ORDER BY total_students DESC
LIMIT 1;
"""
pd.read_sql(query,conn)
```

```python
query="""
SELECT Year, COUNT(*) AS total_students
FROM students
GROUP BY Year
ORDER BY Year;
"""
pd.read_sql(query,conn)
```

```python
query="""
SELECT City, COUNT(*) AS total_students
FROM students
GROUP BY City
ORDER BY total_students DESC
LIMIT 5;
"""
pd.read_sql(query,conn)
```

```python
query = """
SELECT AVG(
    CAST(strftime('%Y','now') AS INTEGER) - 
    CAST(strftime('%Y', Date_of_Birth) AS INTEGER)
) AS avg_age
FROM students;
"""
pd.read_sql(query,conn)
```

```python
query="""
SELECT Course, Gender, COUNT(*) AS total
FROM students
GROUP BY Course, Gender
ORDER BY Course;
"""
pd.read_sql(query,conn)
```
