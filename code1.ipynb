```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import  seaborn as sns
```

```python
df = pd.read_csv("cleaned_student_dataset.csv")
```

```python
#numerical column
df.describe()
```

```python
#categorical columns
print("Gender Distribution:")
print(df['Gender'].value_counts())

print("\nCourse Distribution:")
print(df['Course'].value_counts())

print("\nCity Distribution:")
print(df['City'].value_counts())
```

```python
df['Gender'].value_counts()
df['Course'].value_counts()
```

```python
#bar graph
plt.figure(figsize=(6,4))
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.show()
```

```python
df['Course'].value_counts()
plt.figure(figsize=(8,4))
df['Course'].value_counts().plot(kind='bar')
plt.title("Students per Course")
plt.xticks(rotation=45)
plt.show()
```

```python
df['City'].value_counts().head(10)

plt.figure(figsize=(8,4))
df['City'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Cities by Enrollment")
plt.xticks(rotation=45)
plt.show()
```

```python
#Multivariate Analysis & Correlation
plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Year'])
plt.xlabel("Age")
plt.ylabel("Academic Year")
plt.title("Age vs Academic Year")
plt.show()
```

```python
pd.crosstab(df['Course'], df['Gender']).plot(kind='bar', stacked=True)
plt.xticks(rotation=45)
plt.title("Course vs Gender Distribution")
plt.show()
```

```python
corr = df[['Age','Year']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Age','Year']])
plt.show()
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
total_students = df.shape[0]
avg_age = df['Age'].mean()
top_course = df['Course'].value_counts().idxmax()
top_city = df['City'].value_counts().idxmax()

print("Total Students:", total_students)
print("Average Age:", avg_age)
print("Most Popular Course:", top_course)
print("Top City:", top_city)
```
